```python
import pandas as pd

data = {'x' : [1., 2., 3., 4., 5.], 'y' : [1., 3., 4., 6., 5.]}
df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['x'], df['y'])
plt.grid()
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

X = df[['x']]
y = df['y']

reg = LinearRegression()
reg.fit(X, y)
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error

pred = reg.predict(X)
```

```python
plt.scatter(df['x'], df['y'])
plt.scatter(df['x'], pred)
plt.grid()
plt.show()
```

```python
np.sqrt(mean_absolute_error(y, pred))
```

```python
plt.scatter(y, pred)
plt.plot([0, 6], [0, 6], 'r')
plt.xlabel('Actual')
plt.ylabel('Predict')
plt.title('Real vs Predict')
plt.grid()
plt.show()
```
