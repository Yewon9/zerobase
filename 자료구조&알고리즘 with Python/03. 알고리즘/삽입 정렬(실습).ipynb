### 1부터 1000까지의 난수 100개를 생성하고 다음의 요구사항을 만족하는 모듈을 만들어보자.


```python
class SortNumbers:
    def __init__(self, ns, asc = True):
        self.nums = ns
        self.isAsc = asc

    def isAscending(self, flag):
        self.isAsc = flag

    def setSort(self):
        for i1 in range(1, len(self.nums)):
            i2 = i1 - 1
            cNum = self.nums[i1]

            if self.isAsc:
                while self.nums[i2] > cNum and i2 >= 0:
                    self.nums[i2 + 1] = self.nums[i2]
                    i2 -= 1
            else:
                while self.nums[i2] < cNum and i2 >= 0:
                    self.nums[i2 + 1] = self.nums[i2]
                    i2 -= 1
            self.nums[i2 + 1] = cNum

    def getSortedNumbers(self):
        return self.nums

    def getMinNumber(self):
        if self.isAsc:
            return self.nums[0]
        else:
            return self.nums[len(self.nums) - 1]

    def getMaxNumber(self):
        if self.isAsc:
            return self.nums[len(self.nums) - 1]
        else:
            return self.nums[0]
```

```python
import random
```

```python
nums = random.sample(range(1, 1000), 100)
print(f'not sorted numbers : {nums}')
```

```python
sn = SortNumbers(nums)
```

```python
# 오름차순
sn.setSort()
sortedNumbers = sn.getSortedNumbers()
print(f'sortedNumbers : {sortedNumbers}')
```

```python
# 내림차순
sn.isAscending(False)
sn.setSort()
sortedNumbers = sn.getSortedNumbers()
print(f'sortedNumbers : {sortedNumbers}')
```

```python
# 최대 & 최소
print(f'min : {sn.getMinNumber()}')
print(f'max : {sn.getMaxNumber()}')
```
