1부터 100까지의 정수 중 2의 배수와 3의 배수를 구분해서 출력하자. (단, while문을 이용하자.)

```python
n = 1
while n <= 101:
  if n % 2 == 0:
    print('{}은 2의 배수이다.'.format(n))
  elif n % 3 == 0:
    print('{}은 3의 배수이다.'.format(n))

  n += 1
```

while문을 이용해서 사용자가 입력한 구구단을 출력하자.

```python
gugudanNum = int(input('희망 구구단 입력 : '))

n = 1
while n < 10:
  result = gugudanNum * n
  print('{} * {} = {}'.format(gugudanNum, n, result))
  n += 1
```
